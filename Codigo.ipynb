Issue 1) Importar los Excel para trabajar con ellos

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
Excel_files = ['CERTAMEN1/MAYO.xlsx','CERTAMEN1/JUNIO.xlsx','CERTAMEN1/JULIO.xlsx', 'CERTAMEN1/AGOSTO.xlsx']

dataSetMayo   = pd.read_excel(Excel_files[0], sheet_name=None)
dataSetJunio  = pd.read_excel(Excel_files[1], sheet_name=None)
dataSetJulio  = pd.read_excel(Excel_files[2], sheet_name=None)
dataSetAgosto = pd.read_excel(Excel_files[3], sheet_name=None)
```

```python
for i in range (len(Excel_files)):
    xlsx = pd.ExcelFile(Excel_files[i])
dataSet_sheets = []
for sheet in xlsx.sheet_names:
    dataSet_sheets.append(xlsx.parse(sheet))
    SetMayo = pd.concat(dataSetMayo, sort=False)
    SetJunio = pd.concat(dataSetJunio, sort=False)
    SetJulio = pd.concat(dataSetJulio, sort=False)
    SetAgosto = pd.concat(dataSetAgosto, sort=False)
```

```python
AguaMayo = pd.DataFrame(SetMayo,columns = ['CANATIDAD DE AGUA ','OBSERVACIONES '])
AguaJunio = pd.DataFrame(SetJunio,columns = ['CANATIDAD DE AGUA ','OBSERVACIONES '])
AguaJulio = pd.DataFrame(SetJulio,columns = ['CANATIDAD DE AGUA ','OBSERVACIONES '])
AguaAgosto = pd.DataFrame(SetAgosto,columns = ['CANATIDAD DE AGUA ','OBSERVACIONES '])
```

```python
ObservacionesMayo = AguaMayo.iloc[:,1]
ObsMayo = ObservacionesMayo.drop_duplicates(keep="first")
promMayo=np.arange(25)
for n in range(25):
    promMayo[n]=float(AguaMayo.iloc[:,0][8*n:8*n+8].mean())
promMayo=np.delete(promMayo,5)
promMayo=np.delete(promMayo,10)
promMayo=np.delete(promMayo,15)
promMayo=np.delete(promMayo,21)
promMayo=np.delete(promMayo,20)

SheetsMayo = ObservacionesMayo.index.get_level_values(0)
SheetsMayoUnduplicate = SheetsMayo.drop_duplicates(keep="first")
SheetsMayoUnduplicate=np.delete(SheetsMayoUnduplicate,5)
SheetsMayoUnduplicate=np.delete(SheetsMayoUnduplicate,10)
SheetsMayoUnduplicate=np.delete(SheetsMayoUnduplicate,15)
SheetsMayoUnduplicate=np.delete(SheetsMayoUnduplicate,21)
SheetsMayoUnduplicate=np.delete(SheetsMayoUnduplicate,20)
```

```python
fig = plt.figure(u'Gr치fica de barras',figsize=(20,10)) # Figure
ax = fig.add_subplot(111) # Axes
nombres = SheetsMayoUnduplicate
datos = promMayo
xx = range(len(datos))

ax.bar(xx, datos, width=0.2, align='center')
ax.set_xticks(xx)
ax.set_xticklabels(nombres)
ax.legend(loc="lower right", title="Legend Title")
plt.show()
```
